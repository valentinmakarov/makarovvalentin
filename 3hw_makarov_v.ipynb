 **Статистический машинный перевод (SMT)** - это разновидность машинного перевода, где перевод генерируется на основе статистических моделей.

*Каким должен быть перевод?*

1) Правдоподобным

2) Лингвистически связным


**Задача:**
дано предложения на языке X
необходимо найти самый правдоподобный вариант перевода этого предложения на язык Y


SMT похож на модель зашумленного канала (*"The noisy channel"*), где у нас в процессе передачи некого аудиосигнала X, он через помехи превращается в Y. Соответственно, как восстановить X? Посмотреть, какому Y может соответствовать X

**Компоненты SMT:**

1) ***N-граммная языковая модель*** (P(Y) одноязычная - смотрим на вероятностное распределение конструкций на уровне слов или фраз в языке Y

2) ***Модель перевода:*** (P(X|Y) двуязычная -  ищем статистику соответствий фраз в параллельном корпусе, ищем переводческие соотвествия X-Y, моделируем их с помощью теории вероятности.

Используем T-tables, где в таблице слово y и его условная вероятность t(x|y), смотрим на  *translation probability*.

Предполагаем, что любое предложение на языке Y - это "искаженная версия" некой фразы на языке X, ищем наиболее вероятностные соответствия X-Y

 Выравнивание (*alignment*) - соответствие между X и Y на уровне слов или фраз.

3) ***Модель-декодер*** для преобразования X в Y : все варианты перевода - наши гипотезы, но каждую гипотезу перебирать мы не можем, нужно отобрать наиболее грамматичный и лексически правдопободный вариант перевода. Есть два способа:

1) будем считать лучшим первое совпадение

2) Поиск по графу (*Beam search*)

Но это еще не все :) нужно посмотреть на результаты работы и оценить их, например с помощью **BLEU-score**
